```python
rows = []

for file in tqdm(filenames):
    with open(file, "r") as f:
        data = safe_load(f)

    if isinstance(data, dict):
        rows.append(data)

final_df = pd.json_normalize(rows)
```

```python
bowling_team=[]
ball_of_match=[]
batsman=[]
bowler=[]
runs=[]
player_dismissed=[]
batting_team=[]
match_id=[]
city=[]
id=0
# final_df_copy.iloc[86][0][0]
for index,rows in final_df_copy.iterrows():
    id+=1 
    if id in[2976,3043,4028] :
        continue
    for delivery in rows[0][0]['1st innings']['deliveries']:
        ball=list(delivery.keys())[0]
        match_id.append(id)
        ball_of_match.append(ball)
        bowling_team.append(rows[0][1]['2nd innings']['team'])
        batsman.append(delivery[ball]['batsman'])
        bowler.append(delivery[ball]['bowler'])
        runs.append(delivery[ball]['runs']['total'])
        city.append(rows['info.city'])
        if 'wicket' in delivery[ball]:
            player_dismissed.append(delivery[ball]['wicket']['player_out'])
        else:
            player_dismissed.append(0)
        batting_team.append(rows[0][0]['1st innings']['team'])

delivery_df=pd.DataFrame({'match_id':match_id,
                                  'batting_team':batting_team,
                                  'bowling_team':bowling_team,
                                  'ball_of_match':ball_of_match,
                                  'batsman':batsman,
                                  'bowler':bowler,
                                  'runs':runs,
                                  'player_dismissed':player_dismissed,
                                  'city':city
                                  })
```

```python
top_cities=new_final_df['city'].value_counts()[new_final_df['city'].value_counts()>240].index.tolist()
```

```python
new_final_df=new_final_df[new_final_df['city'].isin(top_cities)]
```

```python
new_final_df['runs'] = pd.to_numeric(new_final_df['runs'])
new_final_df['current_score']=new_final_df.groupby('match_id')['runs'].cumsum()
```

```python
new_final_df.head(20)
```

```python
new_final_df['over']=new_final_df['ball_of_match'].apply(lambda x : str(x).split('.')[0])
new_final_df['ball_no.']=new_final_df['ball_of_match'].apply(lambda x : str(x).split('.')[1])
new_final_df['balls_bowled']=new_final_df['over'].astype(int)*6+new_final_df['ball_no.'].astype(int)
new_final_df['balls_left']=120-new_final_df['balls_bowled']
new_final_df['balls_left']=new_final_df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
new_final_df.tail(20)
```

```python
new_final_df['player_dismissed']=new_final_df['player_dismissed'].apply(lambda x: 0 if x==0 else 1)
```

```python
new_final_df['wickets_left']=10-new_final_df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
new_final_df['crr']=(new_final_df['current_score']*6)/new_final_df['balls_bowled']
```

```python
groups=new_final_df.groupby('match_id')
match_ids=new_final_df['match_id'].unique()
last_five=[]
for id in match_ids:
    last_five.extend(groups.get_group(id)['runs'].rolling(window=30).sum())
new_final_df['last_five'] = last_five
```

```python
new_final_df = new_final_df.groupby('match_id').sum()['runs'].reset_index().merge(new_final_df,on='match_id')
```

```python
new_final_df=new_final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
new_final_df.dropna(inplace=True)
new_final_df.isnull().sum()
```

```python
new_final_df.rename(columns={'runs_x':'final_score'},inplace=True)
```

```python
new_final_df.head(10)
```

```python
required_df = new_final_df.sample(new_final_df.shape[0])
```

```python
required_df.head(10)
```

```python
X = required_df.drop(columns=['final_score'])
y = required_df['final_score']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
